## Scraping [1/2]
Récupération des données

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.spin-off.fr/calendrier_des_series.html?date=2023-10'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

for tr in soup.find_all('tr', class_=False):
    tds = tr.find_all('td', class_='floatleftmobile td_jour td_jourleft')
    for td in tds:
        link_elem = td.find('a', class_='liens')
        if link_elem:
            info = link_elem['alt'].split()
            nom_serie = ' '.join(info[0:-3])
            num_saison = info[-3]
            num_episode = info[-1]
            pays_origine = td.find('img')['alt']
            chaine_diffusion = td.find_all('img')[-1]['alt']
            url_relative = link_elem['href']
            # Trouver la div contenant le jour de diffusion
            date_diffusion_elem = tr.find_previous_sibling('div', class_='div_jour')
            # Extraire le texte du jour de diffusion, s'il existe
            date_diffusion = date_diffusion_elem.find('div', class_='jour_semaine_mobile').text if date_diffusion_elem else 'Date inconnue'

            print(f'Nom de la série: {nom_serie}')
            print(f'Numéro de l’épisode: {num_episode}')
            print(f'Numéro de la saison: {num_saison}')
            print(f'Date de diffusion de l’épisode: {date_diffusion}')
            print(f'Pays d’origine: {pays_origine}')
            print(f'Chaîne qui diffuse la série: {chaine_diffusion}')
            print(f'URL relative de la page de l’épisode: {url_relative}')
```
